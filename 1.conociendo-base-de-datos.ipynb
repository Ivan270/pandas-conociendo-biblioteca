## Importando la base de datos

```python
import pandas as pd

url = "https://gist.githubusercontent.com/ahcamachod/a572cfcc2527046db93101f88011b26e/raw/ffb13f45a79d31223e645611a119397dd127ee3c/alquiler.csv"

pd.read_csv(url)
```

Lo anterior devolverá una tabla mal formateada, porque en este caso el csv no viene con la coma como separador, sino punto y coma. Por lo que habrá que cambiar el delimitador utilizando el parámetro `sep` o en su defecto la versión sin acortar `delimiter`

```python
datos = pd.read_csv(url, delimiter=';')
datos
```

Esto tampoco será la mejor forma de visualizar: Solo se ven las 5 primeras y 5 últimas filas.
Para mejorarlo se puede usar el método `head` de Pandas. 
Devolverá las primeras 5 filas del dataset, para ver las últimas 5 se utiliza el método `tail`. 
Si se quiere ver las primeras diez filas: `datos.head(10)`
Para obtener una muestra aleatoria de 10 filas: `datos.sample(10)`

```python
datos.sample(10)
```
