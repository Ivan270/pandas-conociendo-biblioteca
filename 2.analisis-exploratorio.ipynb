# Valores iniciales
Debajo se encontrará la base de datos

```python
import pandas as pd

url = "https://gist.githubusercontent.com/ahcamachod/a572cfcc2527046db93101f88011b26e/raw/ffb13f45a79d31223e645611a119397dd127ee3c/alquiler.csv"

pd.read_csv(url)

datos = pd.read_csv(url, delimiter=";")
```

## Demandas ML
### Valor promedio de inmuebles por tipo

```python
datos.head()
```

```python
# Promedio de todos los valores. Necesitamos separar por tipo de propiedad!!
datos['Valor'].mean()
```

```python
datos.groupby(['Tipo']).mean(numeric_only=True)
```

```python
# Para mostrar solamente los valores de la columna 'Valor'
# Devolverá los valores promedio dependiendo del tipo de inmueble
datos.groupby('Tipo')['Valor'].mean(numeric_only=True)
```

```python
# Para dejarlo como DataFrame y organizados en orden ascendente, se indica en el argumento qué valor se utilizará para ordenar
datos.groupby("Tipo")[["Valor"]].mean().sort_values('Valor')
```

```python
# Almacenando en variable
df_tipo_precio = datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")

import matplotlib.pyplot as plt
# Para graficar -> kind: tipo de grafico, fiffsize: tamaño figura, color: color
df_tipo_precio.plot(kind='barh', figsize=(12,8), color='purple')
```

## Corrección: Inmuebles residenciales
Habiendo presentado el reporte anterior, los compañeros de ML nos indican que quieren hacer una corrección:
**Solo quieren enfocarse en los inmuebles residenciales**

```python
# Alternativa a slice notation
# Se seleccionarán los valores únicos que hay en la columna 'Tipo'
datos.Tipo.unique()

# Se crea lista con selección de inmuebles comerciales
inmuebles_comerciales = [
    "Conjunto Comercial/Sala",
    "Edificio Completo",
    "Tienda/Salón",
    "Casa Comercial",
    "Terreno Estándar",
    "Cochera/Estacionamiento",
    "Galpón/Depósito/Almacén",
    "Tienda en Centro Comercial",
    "Hotel",
    "Loteo/Condominio",
    "Industria",
]
```

```python
# Consultar si alguno de los elementos de la lista creada inmuebles_comerciales se encuentra dentro del dataset
datos.query('@inmuebles_comerciales in Tipo')
```

```python
# Por el contrario, queremos consultar los elementos que NO ESTÁN presentes en inmuebles comerciales
datos.query('@inmuebles_comerciales not in Tipo')
```

```python
# Almacenar consulta en variable llamada dataframe o df
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
# Volviendo a hacer la comparación y visualización con nuevos datos filtrados

df_tipo_precio = df.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")

import matplotlib.pyplot as plt

# Para graficar -> kind: tipo de grafico, fiffsize: tamaño figura, color: color
df_tipo_precio.plot(kind="barh", figsize=(12, 8), color="purple")
```
